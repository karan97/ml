```python
test=[]
prediction=[]
street = os.listdir('/kaggle/input/intel-image-classification/seg_test/seg_test/street/')
%time
for a in street:
    try:
        image = cv2.imread('/kaggle/input/intel-image-classification/seg_test/seg_test/street/'+a)
        image_to_array=Image.fromarray(image,"RGB")
        size_image=image_to_array.resize(img_size)
        test.append(np.array(size_image))
        prediction.append(0)
    except AttributeError:
        print("f")
        
sea = os.listdir('/kaggle/input/intel-image-classification/seg_test/seg_test/sea/')
%time
for b in sea:
    try:
        image = cv2.imread('/kaggle/input/intel-image-classification/seg_test/seg_test/sea/'+b)
        image_to_array=Image.fromarray(image,"RGB")
        size_image=image_to_array.resize(img_size)
        test.append(np.array(size_image))
        prediction.append(1)
    except AttributeError:
        print("e")
        
glacier = os.listdir('/kaggle/input/intel-image-classification/seg_test/seg_test/glacier/')
%time
for c in glacier:
    try:
        image = cv2.imread('/kaggle/input/intel-image-classification/seg_test/seg_test/glacier/'+c)
        image_to_array=Image.fromarray(image,"RGB")
        size_image=image_to_array.resize(img_size)
        test.append(np.array(size_image))
        prediction.append(2)
    except AttributeError:
        print("d")
        
mountain = os.listdir('/kaggle/input/intel-image-classification/seg_test/seg_test/mountain/')
%time
for d in mountain:
    try:
        image = cv2.imread('/kaggle/input/intel-image-classification/seg_test/seg_test/mountain/'+d)
        image_to_array=Image.fromarray(image,"RGB")
        size_image=image_to_array.resize(img_size)
        test.append(np.array(size_image))
        prediction.append(3)
    except AttributeError:
        print("c")
        
        
buildings = os.listdir('/kaggle/input/intel-image-classification/seg_test/seg_test/buildings/')
%time
for e in buildings:
    try:
        image = cv2.imread('/kaggle/input/intel-image-classification/seg_test/seg_test/buildings/'+e)
        image_to_array=Image.fromarray(image,"RGB")
        size_image=image_to_array.resize(img_size)
        test.append(np.array(size_image))
        prediction.append(4)
    except AttributeError:
        print("b")
        
forest = os.listdir('/kaggle/input/intel-image-classification/seg_test/seg_test/forest/')
%time
for f in forest:
    try:
        image = cv2.imread('/kaggle/input/intel-image-classification/seg_test/seg_test/forest/'+f)
        image_to_array=Image.fromarray(image,"RGB")
        size_image=image_to_array.resize(img_size)
        test.append(np.array(size_image))
        prediction.append(5)
    except AttributeError:
        print("a")
```

```python
x_test=np.array(test)
y_test=np.array(prediction)
y_test.shape
```

```python
import tensorflow as tf
import tensorflow.keras as keras
from keras.utils import to_categorical
no_classes=6
ytr=to_categorical(y_train,no_classes)
yte=to_categorical(y_test,no_classes)
```

```python
from keras.models import Sequential
from keras.layers import  Dense, Dropout, Flatten, Conv2D, MaxPooling2D
model=Sequential()
model.add(Conv2D(filters=32,kernel_size=2,padding="same",activation="relu",input_shape=(128,128,3)))
model.add(MaxPooling2D(pool_size=(4,4)))
model.add(Dropout(0.2))
model.add(Conv2D(filters=32,kernel_size=2,padding="same",activation="relu"))
model.add(MaxPooling2D(pool_size=2))
model.add(Dropout(0.5))
model.add(Conv2D(filters=128,kernel_size=2,padding="same",activation="relu"))
model.add(MaxPooling2D(pool_size=2))
model.add(Dropout(0.7))
model.add(Conv2D(filters=128,kernel_size=2,padding="same",activation="relu"))
model.add(MaxPooling2D(pool_size=2))
model.add(Flatten())
model.add(Dense(500,activation="relu"))
model.add(Dense(6,activation="softmax"))
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
batch_size=500
epochs=80
predict=model.fit(x_train,ytr,batch_size=batch_size,epochs=epochs,validation_data=(x_test,yte))
```

```python
accuracy = model.evaluate(x_test, yte, verbose=1)
print('Test_Accuracy:-', accuracy[1])
```

```python
accuracy = model.evaluate(x_train, ytr, verbose=1)
print('Train_Accuracy:-', accuracy[1])
```
