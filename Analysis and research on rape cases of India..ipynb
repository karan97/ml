```python
from sklearn.cluster import KMeans
```

```python
kmean = KMeans(n_clusters=32)
```

```python
X = rapen.drop(['Subgroup','Area_Name'],axis=1)
```

```python
kmean.fit(X)
```

```python
kmean.cluster_centers_
```

```python
kmean.labels_
```

```python
kmeans = KMeans(n_clusters=35,init='k-means++',max_iter=100,random_state=0) 
y_kmeans = kmeans.fit_predict(X)

#X = X.as_matrix(columns=None)
plt.subplots(figsize=(10,6))

# Visualising the clusters
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0,1],s=100,c='red',label='Andaman')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1,1],s=100,c='blue',label='AndhraPradesh')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2,1],s=100,c='green',label='Arun.pradesh')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3,1],s=100,c='violet',label='Assam')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4,1],s=100,c='indigo',label='Bihar')
plt.scatter(X[y_kmeans == 5, 0], X[y_kmeans == 5,1],s=100,c='brown',label='Chandigarh')
plt.scatter(X[y_kmeans == 6, 0], X[y_kmeans == 6,1],s=100,c='yellow',label='Chattisgarh')
plt.scatter(X[y_kmeans == 7, 0], X[y_kmeans == 7,1],s=100,c='orange',label='Dadra')
plt.scatter(X[y_kmeans == 8, 0], X[y_kmeans == 8,1],s=100,c='black',label='Daman')
plt.scatter(X[y_kmeans == 9, 0], X[y_kmeans == 9,1],s=100,c='green',label='Delhi')
plt.scatter(X[y_kmeans == 10, 0], X[y_kmeans == 10,1],s=100,c='pink',label='Goa')
plt.scatter(X[y_kmeans == 11, 0], X[y_kmeans == 11,1],s=100,c='Tan',label='Gujrat')
plt.scatter(X[y_kmeans == 12, 0], X[y_kmeans == 12,1],s=100,c='purple',label='Haryana')
plt.scatter(X[y_kmeans == 13, 0], X[y_kmeans == 13,1],s=100,c='Chocolate',label='Himachal')
plt.scatter(X[y_kmeans == 14, 0], X[y_kmeans == 14,1],s=100,c='Silver',label='Jammu')
plt.scatter(X[y_kmeans == 15, 0], X[y_kmeans == 15,1],s=100,c='Aquamarine',label='Jharkhand')
plt.scatter(X[y_kmeans == 16, 0], X[y_kmeans == 16,1],s=100,c='Maroon',label='Karnatka')
plt.scatter(X[y_kmeans == 17, 0], X[y_kmeans == 17,1],s=100,c='Salmon',label='Kerla')
plt.scatter(X[y_kmeans == 18, 0], X[y_kmeans == 18,1],s=100,c='plum',label='Lakshdweep')
plt.scatter(X[y_kmeans == 19, 0], X[y_kmeans == 19,1],s=100,c='Mistyrose',label='MP')
plt.scatter(X[y_kmeans == 20, 0], X[y_kmeans == 20,1],s=100,c='Lime',label='Maharashtra')
plt.scatter(X[y_kmeans == 21, 0], X[y_kmeans == 21,1],s=100,c='Teal',label='Manipur')
plt.scatter(X[y_kmeans == 22, 0], X[y_kmeans == 22,1],s=100,c='coral',label='Meghalaya')
plt.scatter(X[y_kmeans == 23, 0], X[y_kmeans == 23,1],s=100,c='crimson',label='Mizoram')
plt.scatter(X[y_kmeans == 24, 0], X[y_kmeans == 24,1],s=100,c='Fuchsia',label='Nagaland')
plt.scatter(X[y_kmeans == 25, 0], X[y_kmeans == 25,1],s=100,c='cyan',label='Orissa')
plt.scatter(X[y_kmeans == 26, 0], X[y_kmeans == 26,1],s=100,c='blue',label='Puducherry')
plt.scatter(X[y_kmeans == 27, 0], X[y_kmeans == 27,1],s=100,c='orange',label='Punjab')
plt.scatter(X[y_kmeans == 28, 0], X[y_kmeans == 28,1],s=100,c='green',label='Rajasthan')
plt.scatter(X[y_kmeans == 29, 0], X[y_kmeans == 29,1],s=100,c='red',label='Sikkim')
plt.scatter(X[y_kmeans == 30, 0], X[y_kmeans == 30,1],s=100,c='black',label='TamilNadu')
plt.scatter(X[y_kmeans == 31, 0], X[y_kmeans == 31,1],s=100,c='brown',label='Tripura')
plt.scatter(X[y_kmeans == 32, 0], X[y_kmeans == 32,1],s=100,c='violet',label='Up')
plt.scatter(X[y_kmeans == 33, 0], X[y_kmeans == 33,1],s=100,c='indigo',label='Uttrakhand')
plt.scatter(X[y_kmeans == 34, 0], X[y_kmeans == 34,1],s=100,c='pink',label='West Bengal')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='yellow',label='Centroids')
plt.title('Clusters of RAPE in different places across India')
plt.legend()
plt.show()
```
