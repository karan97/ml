```python
def team1_vs_team2(team1,team2):
    m1=match[((match['team1']==team1)|(match['team2']==team1))&((match['team1']==team2)|(match['team2']==team2))]
    sns.countplot(x='season', hue='winner',data=m1,palette='Set3')
    plt.xticks(rotation='vertical')
    leg = plt.legend( loc = 'upper center')
    fig=plt.gcf()
    fig.set_size_inches(10,6)
    plt.show()
team1_vs_team2('KKR','MI')
```

```python
def comparator(team1):
    teams=['MI','KKR','RCB','DC','CSK','RR','DD','GL','KXIP','SRH','RPS','KTK','PW']
    teams.remove(team1)
    opponents=teams.copy()
    m1=match[((match['team1']==team1)|(match['team2']==team1))]
    for i in opponents:
        mask = (((m1['team1']==i)|(m1['team2']==i)))&((m1['team1']==team1)|(m1['team2']==team1))
        mt2 = m1.loc[mask, 'winner'].value_counts().to_frame()
        print(mt2)
```
